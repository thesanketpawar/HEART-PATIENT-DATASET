```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/heart_dataset.csv')
```

```python
data.head()
```

```python
x=data.drop(['target'],axis=1)
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lm=LogisticRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
```

```python
print(cm)
```

```python
print(ac)
```

PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=13)   #Transforming X train to X Reduced Train and X Reduced Test because it has many features available
x_reduced_train=pd.DataFrame(pca.fit_transform(x_train))  #FIT_TRANSFORM FUNCTION USED TO REDUCE THE VARIABLE 
x_reduced_test=pd.DataFrame(pca.fit_transform(x_test))
```

```python
x_train.shape
```

```python
x_reduced_train.shape
```

```python
x_test.shape
x_reduced_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
```

```python
log_model.fit(x_reduced_train,y_train)
```

```python
new_y_pred=log_model.predict(x_reduced_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm=confusion_matrix(y_test,new_y_pred)
ac=accuracy_score(y_test,new_y_pred)
```

```python
print(cm)
```
